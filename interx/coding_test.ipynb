### 라이브러리

```python
import random
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
```

### 파라미터

```python
data_path = './wine.csv'
```

### 데이터 읽기

```python
wine_df = pd.read_csv(data_path, names=[f'featrue{i}'for i in range(12)]+['label'])
wine_df.head(5)
```

### 임의의 feature 값 추가

```python
add_feature = [random.randint(1,100) for _ in range(6497)] 
wine_df.insert(12, 'feature12', add_feature)
wine_df
```

### 데이터 스케일링

```python
wine_df.describe()
```

```python
# x data
x_data = wine_df.iloc[:, :-1].copy()
x_data.head(3)
```

```python
# y data
y_data = wine_df['label'].values
```

```python
scaler = MinMaxScaler().fit(x_data)
output = scaler.transform(x_data)
prepro_x = pd.DataFrame(output, columns=x_data.columns, index=list(x_data.index.values))
prepro_x.head(3)
```

```python
onehot = OneHotEncoder()
prepro_y = onehot.fit_transform(y_data.reshape(-1, 1))
prepro_y.toarray()
```
